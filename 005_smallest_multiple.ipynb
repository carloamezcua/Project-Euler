**Problem 5: Smallest Multiple**

$2520$ is the smallest number that can be divided by each of the numbers from $1$ to $10$ without any remainder.

What is the smallest positive number that is evenly divisible by all of the numbers from $1$ to $20$?

```python
from euler_lib import factores_primos, potencias_factores
```

```python
max_factores = {}

for i in range(1, 21):
    factores = factores_primos(i)
    potencias_i = potencias_factores(factores)
    for f in potencias_i:
        if f not in max_factores:
            max_factores[f] = potencias_i[f]
        else:
            if potencias_i[f] > max_factores[f]:
                max_factores[f] = potencias_i[f]

resultado = 1
for f in max_factores:
    resultado *= f ** max_factores[f]

print(resultado)
```
