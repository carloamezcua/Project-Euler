**Problem 4: Largest Palindrome Product**

A palindromic number reads the same both ways. The largest palindrome made from the product of two 2-digit numbers is $9009 = 91 \times 99$.

Find the largest palindrome made from the product of two 3-digit numbers.

```python
multiplicando = [9, 0, 0]
multiplicador = [9, 0, 0]
palindromo = []
for decenas_multiplicando in range(10):
    multiplicando[1] = decenas_multiplicando
    for unidades_multiplicando in range(10):
        multiplicando[2] = unidades_multiplicando
        for decenas_multiplicador in range(10):
            multiplicador[1] = decenas_multiplicador
            for unidades_multiplicador in range(10):
                multiplicador[2] = unidades_multiplicador
                MULTIPLICANDO = int(str(multiplicando[0]) +
                                    str(multiplicando[1]) +
                                    str(multiplicando[2]))
                MULTIPLICADOR = int(str(multiplicador[0]) +
                                    str(multiplicador[1]) +
                                    str(multiplicador[2]))
                resultado = MULTIPLICANDO * MULTIPLICADOR
                resultado_str = str(resultado)
                if resultado_str == resultado_str[::-1]:
                    palindromo.append(resultado)
print(max(palindromo))
```
