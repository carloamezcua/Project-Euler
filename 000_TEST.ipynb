```python
n = 21
primos = []
es_primo = [True] * (n + 1)
es_primo[0] = es_primo[1] = False # 0 y 1 no son primos
print(es_primo[0], es_primo[1])  # Verificar que 0 y 1 están marcados como no primos
for i in range(2, n + 1):
    print(f"i: {i}, es_primo[i]: {es_primo[i]}")
    if es_primo[i]:
        primos.append(i)
        for j in range(i * i, n + 1, i): # Marcar los múltiplos de i como no primos
            print(f"  Marcando {j} como no primo (múltiplo de {i})")
            es_primo[j] = False

print(primos)
```
