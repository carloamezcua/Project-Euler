```python
def minimo_comun_multiplo(a, b):
    if a == 0 or b == 0:
        return 0
    else:
        maximo_comun_divisor = maximo_comun_divisor(a, b)
        return abs(a * b) // maximo_comun_divisor
```

```python
def maximo_comun_divisor(a, b):
    while b:
        a, b = b, a % b
    return abs(a)
```

```python
def factores_primos(n):
    factores = []
    divisor = 2
    while divisor <= n:
        if n % divisor == 0:
            factores.append(divisor)
            n //= divisor
        else:
            divisor += 1
    return factores
```
