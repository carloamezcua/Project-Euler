Problem 3:
The prime factors of 13195 are 5, 7, 13 and 29.
What is the largest prime factor of the number 600851475143?

```python
factores_primos = []
n = 600851475143
primo = 2
while n > 1:
    if n % primo == 0:
        factores_primos.append(primo)
        n = n/primo
    else:
        primo += 1
print(factores_primos[-1])
```
