**Problem 3: Largest Prime Factor**

The prime factors of $13195$ are $5, 7, 13$ and $29$.

What is the largest prime factor of the number $600851475143$?

```python
def factores_primos(n):
    factores = []
    divisor = 2
    while divisor <= n:
        if n % divisor == 0:
            factores.append(divisor)
            n //= divisor
        else:
            divisor += 1
    return factores
```
